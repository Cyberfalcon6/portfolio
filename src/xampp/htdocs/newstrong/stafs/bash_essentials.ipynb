## ðŸ“¦ Basic Navigation Commands

| Command             | Description                                |
|:--------------------|:--------------------------------------------|
| `pwd`               | Print **current working directory**        |
| `ls`                | List files and folders                     |

## ðŸ“¦ File & Directory Management
